```python
# MULTIPLE LINEAR REGRESSION
# HOUSE PRICE PREDICTION
```

```python
# import libraries
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
# Create Dataset
```

```python
data={
    'Area':[1200,1500,1800,2400,3000,3500,4000,4500,5000,5500],
    'Bedrooms':[2,3,3,4,5,5,6,6,7,7],
    'Age':[10,5,8,2,1,3,2,4,1,2],
    'Distance':[5,3,4,2,1,2,1,3,2,1],
    'Price':[250,320,340,450,600,650,700,720,800,850]
}
#create dataframe
df=pd.DataFrame(data)
# display dataset
print("\nDATASET\n")
print(df)
```

```python
# prepare input and output
```

```python
# Independent variables
x=df[['Area','Bedrooms','Age','Distance']]
# Dependent variable
y=df['Price']
```

```python
# Build Model
```

```python
model = LinearRegression()
```

```python
# Train model
model.fit(x,y)
```

```python
# Predictions
```

```python
y_pred=model.predict(x)
```

```python
# Model Parameter
```

```python
print("\nMODEL PARAMETERS\n")
print("Intercept:")
print(model.intercept_)
print("\nCoefficients:")
for feature,coef in zip(x.columns,model.coef_):
    print(feature,":",coef)
```

```python
# Regression Equation
```

```python
print("\nREGRESSION EQUATION\n")
print("Price = ", end="")
for feature,coef in zip(x.columns,model.coef_):
    print(f"({coef:.4f}*{feature}) + ", end="")
print(model.intercept_)
```

```python
# Evaluation Metrics
mse=mean_squared_error(y,y_pred)
r2=r2_score(y,y_pred)
print("\nEVALUATION METRICS\n")
print("Mean Squared Error(MSE):",round(mse,4))
print("R-squared score:",round(r2,4))
```

```python
#Predict New Hose Price
```

```python
new_house = pd.DataFrame({'Area':[2800],'Bedroom':[4],'Age':[2],'Distance':[2]})
```

```python
new_house = new_house.rename(columns={'Bedroom': 'Bedrooms'})
predicted_price = model.predict(new_house)
print("\nNEW HOUSE PRICE PREDICTION\n")
print("Predicted House Price:",round(predicted_price[0],2))
```

```python
# Visualization
```

```python
plt.figure(figsize=(8,6))
plt.scatter(y,y_pred,color='blue')
plt.plot([y.min(),y.max()],[y.min(),y.max()],color='red',linestyle='-')
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual Prices vs Predicted Prices')
plt.grid(True)
plt.show()
```
